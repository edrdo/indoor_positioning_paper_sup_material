```python
rssi_train_dataset = pd.read_csv('datasets/brssi/train.tsv',sep='\t')
rssi_train_dataset
```

```python
rssi_test_dataset = pd.read_csv('datasets/brssi/TS.tsv',sep='\t')
rssi_test_dataset
```

## Video processing 

```python
def process_video_file(filename, dataset_tsv_file):
    label = '_'.join(os.path.splitext(os.path.basename(filename))[0].split('_')[1:])
    label = LABEL_CONV[label]
    source = 'Pixel' if 'Pixel' in filename else 'Redmi'
    train_times = rssi_train_dataset[(rssi_train_dataset.label == label) & (rssi_train_dataset.source == source)]['time'].sort_values()
    test_times = rssi_test_dataset[(rssi_test_dataset.label == label) & (rssi_test_dataset.source == source)]['time'].sort_values()
    time_list = [ ('train', t) for t in train_times] + [ ('test', t) for t in test_times ]
    time_list.sort(key=lambda x: x[1])
    vc = VideoCapture.VideoInput(filename)
    print('===>', filename, label, source)
    print("FPS", vc.fps)
    print('Number of frames:', vc.frame_count)
    print('Time:', vc.frame_count / vc.fps, time_list[len(time_list)-1][1])
    for (dt, t) in time_list:
        r, frame = vc.next_frame()
        if not r:
            break
        #frame = cv2.rotate(frame, cv2.ROTATE_90_CLOCKWISE)
        os.makedirs('datasets/vision/images/%s' % label, exist_ok=True)
        image_file = "datasets/vision/images/%s/%s_%s_%04d_%04d.jpeg" % (label, dt,source, t, vc.current_frame)
        cv2.imwrite(image_file, frame)
        print(label, dt, t, source, image_file, filename, sep='\t', file=dataset_tsv_file)
        #print(label, dt, t, source, image_file)
        t += 1   
        while vc.current_time < t:
            _,frame = vc.next_frame()
```

```python
INPUT_DIRS = [
    'raw_data/Pixel/Vision/',
    'raw_data/Redmi/Vision/'
]

EXCLUDE_PATTERNS = [
    'atrio_entre',
    'teste_final',
    '.gitattributes'
]
with open('datasets/vision/vision_dataset.tsv','w') as dataset_tsv_file:
    video_files = sorted([ os.path.join(d,f) for d in INPUT_DIRS for f in os.listdir(d) \
      if [ p for p in EXCLUDE_PATTERNS if p in f ] == []
    ])
    for f in video_files:
        print(f)
        process_video_file(f, dataset_tsv_file)
```
